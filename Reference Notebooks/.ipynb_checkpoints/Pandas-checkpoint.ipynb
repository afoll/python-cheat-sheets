# Pandas

- Pandas is a open-source library with data structures and data manipulation tools.
- The name is derived from the term "panel data"

```python
# pd is the conventional abbreviation assigned during the import statement

import pandas as pd
```

## Pandas Series

```python
# Series are one-dimensional arrays

numbers = range(1,50,5)
pd.Series(numbers)

# note:  left-hand-most column is just the index
```

```python
# note the dtype indicates it as an integer and hte computer is allocating 64bps to each element
```

```python
# create an object series
string = "Hi!", "How", "are", "you", "?"
pd.Series(string)
```

```python
# Note that a series can not have multiple data stypes across a column
```

```python
# Woe unto python's OG dev team:  you can arbitrarily assign an index
# You can also do the same thing with a {dictionary}

marks = [60, 89, 74, 86]

subject = ["Maths", "Science", "English", "Social Science"]

marks_series = pd.Series(marks, index = subject)
marks_series
```

## Manipulating Series

```python
marks_series[marks_series > 75]
```

```python
# change a value
marks_series["Science"] = 99
marks_series
```

```python
# Sorting
import numpy as np

values = pd.Series([23, 45, np.nan, 41, 23, 34, 55, np.nan, 34, 20])
values
```

```python
# Note that sorting does not change the index of the series
values.sort_values(ascending = True)
```

```python
# You can also assign ranks
marks_series.rank(ascending = True)
```
